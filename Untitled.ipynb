```python

# import libraries
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import classification_report, confusion_matrix
import tensorflow
from tensorflow.keras import layers
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
import seaborn as sns
from tensorflow import keras





# Load a single file as a numpy array with allow_pickle=True
def load_file(filepath):
    data = np.load(filepath, allow_pickle=True)
    return data

def cf(y_true, y_pred, fname, title):
    """
    Create a confusion matrix plot save it into a file
    Inputs:
    y_true: actual labels
    y_pred: prediction labels
    fname: file name
    title: title of the plot
    Outputs:
    None
    """
    cm = confusion_matrix(y_true, y_pred, labels=np.unique(y_true))
    cm_sum = np.sum(cm, axis=1, keepdims=True)
    cm_perc = cm / cm_sum.astype(float) * 100
    annot = np.empty_like(cm).astype(str)
    nrows, ncols = cm.shape
    for i in range(nrows):
        for j in range(ncols):
            c = cm[i, j]
            p = cm_perc[i, j]
            if i == j:
                s = cm_sum[i]
                annot[i, j] = f'{p:.1f}%\n{c}/{s}'
            elif c == 0:
                annot[i, j] = '0'
            else:
                annot[i, j] = f'{p:.1f}%\n{c}'

    cm_df = pd.DataFrame(cm, index=np.unique(y_true), columns=np.unique(y_true))
    cm_df.index.name = 'Actual'
    cm_df.columns.name = 'Predicted'
    ax = sns.heatmap(cm_df, cmap="Blues", annot=annot, fmt='')
    ax.set_title(title)
    ax.xaxis.set_ticklabels(cm_df.columns, rotation=45)
    ax.yaxis.set_ticklabels(cm_df.index, rotation=45)
    plt.savefig(fname, dpi=300)
    plt.show()
```

```python
import os
from sklearn.model_selection import train_test_split



def load_processed_npy_files(folder_path):
    data_list = []
    for filename in os.listdir(folder_path):
        if filename.endswith(".npy"):
            file_path = os.path.join(folder_path, filename)
            data = np.load(file_path)
            data_list.append(data)
    return np.concatenate(data_list, axis=0)

# Load processed network data
processed_network_folder = "C:/Users/shrey/Downloads/CS_PROJ/Processed_Network_OO"
data_net = load_processed_npy_files(processed_network_folder)
# Assuming data_net is a NumPy array
num_rows = data_net.shape[0] - (data_net.shape[0] % 10)
data_net_trimmed = data_net[:num_rows]

# Now reshape
data_net_reshaped = np.reshape(data_net_trimmed, (num_rows // 10, 10, data_net.shape[1]))




# Load sensor data
data_swat1 = pd.read_csv("C:/Users/shrey/Downloads/CS_PROJ/Physical_O/SWaT_Dataset_Normal_v1.csv")
data_swat2 = pd.read_csv("C:/Users/shrey/Downloads/CS_PROJ/Physical_O/SWaT_Dataset_Normal_v0.csv")
data_swat3 = pd.read_csv("C:/Users/shrey/Downloads/CS_PROJ/Physical_O/SWaT_Dataset_Attack_v0.csv")  
data_swat = pd.concat([data_swat1, data_swat2]).sample(frac=1).reset_index(drop=True)

# Extract the labels
y_swat = data_swat["Normal/Attack"].replace({'Normal': 0, 'Attack': 1}).astype("int64")
data_swat = data_swat.drop(['Normal/Attack', ' Timestamp'], axis=1).to_numpy()

# Make sure the number of samples in network and sensor data are the same
# Adjust if necessary
data_swat = data_swat[:data_net.shape[0]]

# Prepare the dataset
data_combined = list(zip(data_net_reshaped, data_swat))

# Split the data
X_val, X_test, y_val, y_test = train_test_split(data_combined, y_swat, stratify=y_swat, test_size=0.2)
X_train, X_val, y_train, y_val = train_test_split(X_val, y_val, stratify=y_val, test_size=0.1)

# Extract network and sensor data to feed to the models
X_train_net, X_train_sen = zip(*X_train)
X_test_net, X_test_sen = zip(*X_test)
X_val_net, X_val_sen = zip(*X_val)

# Convert final data to arrays
X_train_net = np.array(X_train_net)
X_test_net = np.array(X_test_net)
X_train_sen = np.array(X_train_sen)
X_test_sen = np.array(X_test_sen)
X_val_sen = np.array(X_val_sen)
X_val_net = np.array(X_val_net)

# One-hot vectoring the labels
y_train_oh = to_categorical(y_train, num_classes=2)  
y_test_oh = to_categorical(y_test, num_classes=2)
y_val_oh = to_categorical(y_val, num_classes=2)

# Proposed deep multimodal model
# Sensor model
sens_inp = layers.Input(shape=(X_train_sen.shape[1],), name="Sensor") 
x = layers.Dense(512, activation='relu')(sens_inp)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dense(256, activation='relu')(x)
sens_out = layers.Dense(32, activation='relu')(x)

# Network model
net_inp = layers.Input(shape=(X_train_net.shape[1], X_train_net.shape[2]), name="Network")
x = layers.LSTM(1024, return_sequences=True)(net_inp)
x = layers.LSTM(256, return_sequences=True)(x)
net_out = layers.LSTM(128)(x)

# Fusion model
x = layers.concatenate([sens_out, net_out])
x = layers.Dense(128, activation='relu')(x)
x = layers.Dense(32, activation='relu')(x)

# Classification layer
out = layers.Dense(2, activation='softmax')(x)

# Final model
model = Model(inputs=[net_inp, sens_inp], outputs=out)
model.summary()

# Callbacks
model_name = "best_model.keras"
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=70)
monitor = keras.callbacks.ModelCheckpoint(model_name, monitor='val_accuracy', verbose=2, save_best_only=True, save_weights_only=False, mode='max')

# Compile and fit the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit([X_train_net, X_train_sen], y_train_oh, epochs=3, batch_size=512, validation_data=([X_val_net, X_val_sen], y_val_oh), callbacks=[early_stop, monitor], verbose=2)

# Get the prediction on test data
y_pred = np.argmax(model.predict([X_test_net, X_test_sen]), axis=1)

# Create and save a confusion matrix
cf(y_test, y_pred, "mm_cf", "Multimodal Model")

# Print out the performance metrics
print(classification_report(y_test, y_pred))
```

```python
#coverting intial csv into npyimport pandas as pd
import numpy as np
import os

def process_batch_and_save(filepaths, batch_output_path):
    max_columns = 0
    combined_data = []

    # Determine the maximum number of columns in any file in the batch
    for filepath in filepaths:
        try:
            data = pd.read_csv(filepath, nrows=1)
            max_columns = max(max_columns, data.shape[1])
        except Exception as e:
            print(f"Error processing file {filepath} during column check: {e}")

    # Process each file in the batch
    for filepath in filepaths:
        try:
            data = pd.read_csv(filepath)

            # Pad with NaNs if necessary to match max_columns
            if data.shape[1] < max_columns:
                padding = pd.DataFrame(np.nan, index=np.arange(len(data)), columns=np.arange(data.shape[1], max_columns))
                data = pd.concat([data, padding], axis=1)

            if not data.empty:
                combined_data.append(data.to_numpy())
            else:
                print(f"Warning: {filepath} is empty.")
        except Exception as e:
            print(f"Error processing file {filepath}: {e}")
   
    if combined_data:
        # Concatenate the arrays
        batch_data = np.concatenate(combined_data, axis=0)
        np.save(batch_output_path, batch_data)

def create_npy_from_csv_in_batches(folder_path, output_file_path, batch_size=40):
    all_filenames = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
    total_files = len(all_filenames)
    batch_files = []

    for i in range(0, total_files, batch_size):
        batch_filenames = all_filenames[i:i+batch_size]
        filepaths = [os.path.join(folder_path, f) for f in batch_filenames]
        print(f"Processing batch {i//batch_size + 1}/{(total_files - 1)//batch_size + 1}")
        batch_output_path = f"{output_file_path}_batch_{i//batch_size + 1}.npy"
        batch_files.append(batch_output_path)
        process_batch_and_save(filepaths, batch_output_path)

    # Start with an empty array
    combined_data = np.array([])
   
    # Incrementally load each batch file and concatenate
    for batch_file in batch_files:
        batch_data = np.load(batch_file, allow_pickle=True)
        if combined_data.size == 0:
            combined_data = batch_data
        else:
            combined_data = np.concatenate((combined_data, batch_data), axis=0)
        os.remove(batch_file)  # Delete the batch file to free up disk space

    # Save the final combined data
    np.save(output_file_path, combined_data)

# Replace with your actual folder paths and desired output file paths
network_folder_path = '/home/student/Downloads/SWAT/Network'
output_npy_file = '/home/student/Downloads/SWAT/Network_O/Network.npy'

# Uncomment the line below to run the function with your file paths
# create_npy_from_csv_in_batches(network_folder_path, output_npy_file, batch_size=40)
```

```python
#preprocessing the npy files
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.impute import SimpleImputer
import os

def process_file(filepath, chunk_size, labelencoder, imputer, scaler, output_folder):
    """
    Process a single npy file and save the processed chunks to the disk.
    """
    data = np.load(filepath, allow_pickle=True)
    data = np.delete(data, [0, 1, 2, 3], axis=1)

    for i in range(0, len(data), chunk_size):
        chunk = data[i:i + chunk_size]
        df = pd.DataFrame(chunk)
        
        if 13 in df.columns:
            df.drop(13, axis=1, inplace=True)

        # Convert non-numeric columns to strings before label encoding
        for col in df.columns:
            if df[col].dtype == object or df[col].dtype == 'O':
                df[col] = df[col].astype(str)
                df[col] = labelencoder.fit_transform(df[col])
            elif df[col].dtype != np.number:
                df[col] = df[col].apply(lambda x: str(x) if not pd.isnull(x) else x)
                df[col] = labelencoder.fit_transform(df[col])

        df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
        df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)

        chunk_filename = f"processed_chunk_{i // chunk_size}.npy"
        np.save(os.path.join(output_folder, chunk_filename), df.values)

labelencoder = LabelEncoder()
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
scaler = MinMaxScaler()

folder_path = "/home/student/Downloads/SWAT/Network_O"
output_folder = "/home/student/Downloads/SWAT/Processed_Network_OO"
chunk_size = 20000

for filename in os.listdir(folder_path):
    if filename.endswith(".npy"):
        filepath = os.path.join(folder_path, filename)
        print(f"Processing file: {filename}")
        process_file(filepath, chunk_size, labelencoder, imputer, scaler, output_folder)
```
